```python
from collections import defaultdict
def frequent_words(text, k):
    #create a frequent map to count the number of each k-mer
    
    frequency_map = defaultdict(int)
    for i in range(len(text)- k + 1):
        k_mer = text[i: i+k]
        frequency_map[k_mer] += 1
        
    # sort the frequency map by the count of each k-mer, in descending order
    
    sorted_frequency_map = sorted(frequency_map.items(), key = lambda x: x[1] , reverse = True) # list of tuple

    most_common = []

    for element in sorted_frequency_map:
        if element[1] == sorted_frequency_map[0][1]: #element[1] : count of k_mer, element[0]: k_mer
            most_common.append(element[0])
        else:
            break
    return most_common
```

#### 1.4 Solve the Clump Finding Problem


The Clump Finding Problem is a computational biology problem that involves identifying patterns in a string, called a genome, that appear in close proximity to one another (forming "clumps"). The input to the problem includes a genome string, an integer k (representing the length of the patterns to be searched for), an integer L (representing the window size in which the patterns must appear), and an integer t (representing the minimum number of times a pattern must appear within the window to be considered a "clump"). The output of the problem is a list of all distinct k-mers (substrings of length k) that form (L, t)-clumps in the genome.

**request**: search in the windows of L length, get the k_mers with count > t times  

```python
def ClumpFinding(genome, k, L, t):
    clumps = set()
    for i in range(len(genome)-L +1):
        window = genome[i:i+L]
        kmer_counts = {}
        
        for j in range(len(window)- k + 1):
            kmer = window[j:j+k]
            if kmer in kmer_counts:
                kmer_counts[kmer] += 1
            else:
                kmer_counts[kmer] = 1
        
        for kmer, count in kmer_counts.items():
            if count >= t:
                clumps.add(kmer)
    print (" ".join(list(clumps)))
```

#### 1.7 Minimum Skew problem

Skewi+1(Genome) from Skewi(Genome) according to the nucleotide in position i of Genome. 

- If this nucleotide is G, then Skewi+1(Genome) = Skewi(Genome) + 1;
- if this nucleotide is C, then Skewi+1(Genome)= Skewi(Genome) – 1;
- otherwise, Skewi+1(Genome) = Skewi(Genome).

```python
def MinimumSkew(genome):
    skew = [0] * (len(genome) + 1)
    for i in range(1, len(genome)+1):
        if genome[i-1] == 'G':
            skew[i] = skew[i-1] + 1
        elif genome[i-1]  == 'C':
            skew[i] = skew[i-1] -1
        else:
            skew[i] = skew[i-1]
    min_skew = min(skew)
    min_skew_positions = [str(i) for i in range(len(skew)) if skew[i] == min_skew]

    return " ".join(min_skew_positions)
```

#### 1.8 Hamming Distance Problem




```python
def HammingDistance(genome1, genome2):
    if len(genome1) != len(genome2):
        return None
    hammingDistance = 0
    for i in range(len(genome1)):
        if genome1[i] != genome2[i]:
            hammingDistance += 1
    return hammingDistance 

            
```

#### 1.8 Approximate Pattern Matching Problem



```python
def approximate_pattern_matching(k_mer, genome, d):
    n = len(genome)
    m = len(k_mer)
    result = []
    count = 0
    
    for i in range(n - m + 1):
        mismatch = 0
        for j in range(m):
            if k_mer[j] != genome[i+j]:
                mismatch += 1
                if mismatch > d:
                    break
        if mismatch <= d:
            result.append(i)
            count += 1
    result = [str(item) for item in result ]
    return " ".join(result), count
    
```
